## Приоритизация гипотез

В рамках дальнейшего анализа и исследования нам необходимо произвести приоритизацию гипотез для отбора в тестирование наиболее перспективных идей. Для этого мы будем применять  фреймворки ICE и RICE с помощью заданных значений параметров. Нам известно, что в файле `hypothesis.csv` 9 гипотез по увеличению выручки интернет-магазина с указанными параметрами Reach, Impact, Confidence, Effort.

### Приоритизация гипотез по фреймворку ICE 

Применим фреймворк ICE для приоритизации гипотез. Отсортируем их по убыванию приоритета.

```python
hypothesis['ICE'] = round(((hypothesis['Impact'] * hypothesis['Confidence'])/hypothesis['Efforts']),2)

hypothesis[['Hypothesis','ICE']].sort_values(by='ICE',ascending=False)
```

Абсолютным фаворитом стала гипотеза `Запустить акцию, дающую скидку на товар в день рождения` набрав значение 16.2 

2 и 3 место по приоритетности заняли гипотезы: `Добавить два новых канала привлечения трафика, что позволит привлекать на 30% больше пользователей	` и `Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок` набрав 13.3 и 11.2 соответственно.

### Приоритизация гипотез по фреймворку RICE 

Применим фреймворк RICE для приоритизации гипотез. Отсортируем их по убыванию приоритета.

```python
hypothesis['RICE'] = (hypothesis['Impact'] * hypothesis['Confidence'] * hypothesis['Reach'] )/hypothesis['Efforts']

hypothesis.sort_values(by='RICE',ascending=False)
```

Абсолютным фаворитом стала гипотеза `Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок` набрав значение 112

2 и 3 место по приоритетности заняли гипотезы: `Добавить блоки рекомендаций товаров на сайт интернет магазина, чтобы повысить конверсию и средний чек заказа` и `Добавить два новых канала привлечения трафика, что позволит привлекать на 30% больше пользователей` набрав 56 и 40 соответственно.

### Изменение приоритизации гипотез при применении RICE вместо ICE

Укажем, как изменилась приоритизация гипотез при применении RICE вместо ICE. Объясним, почему так произошло.

После применения фреймворка RICE на лидирующие позиции вышли гипотезы: 
- `Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок`
- `Добавить блоки рекомендаций товаров на сайт интернет магазина, чтобы повысить конверсию и средний чек заказа`

Гипотеза `Запустить акцию, дающую скидку на товар в день рождения`, которая являлась самой приоритетной по фреймворку ICE, отошла на 5-ое место по приоритизации.

Параметр **Reach**  учитывает скольких пользователей затронет изменение, которое мы хотим внести, данный параметр используется только в фреймворке RICE, соответственно дополнив фреймворк новым параметром мы изменили алгоритм расчёта приоритизации.

**Вывод**

От выбора способа приоритизации гипотез будет зависеть, то какие гипотезы окажутся приоритетными.

- Добавить два новых канала привлечения трафика, что позволит привлекать на 30% больше пользователей
- Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок

Лидерами по  фреймворку **RICE** оказались гипотезы:

## Анализ A/B-теста

По отобранной гипотезе мы провели А/B-тест. Весь трафик был поделён на две группы: А(контрольная) и B(экспериментальная). A/B-тестирование позволит узнать различие между группами или его отсутствие. 

В рамках анализа вычислим:
- Статистическую значимость различий в среднем числе заказов на посетителям групп A и B по «сырым» и «очищенным»  данным
- Статистическую значимость различий в среднем чеке заказа на посетителям групп A и B по «сырым» и «очищенным»  данным

В рамках анализа построим графики и проанализируем различия:
- кумулятивной выручки групп A и B
- кумулятивного среднего чека групп A и B
- относительного изменения кумулятивного среднего чека группы B к группе A
- кумулятивного среднего количества заказов на посетителя групп A и B
- относительного изменения кумулятивного среднего количества заказов на посетителя группы B к группе A

Исследуем выбросы и аномалия влияющие на результат тестов.

### Подготовка к анализу A/B-теста

Перед анализом полученных данных, нам следует убедится, что
- Количество посетителей в группах различается не более чем на 1%
- Попавший в одну из групп пользователь остаётся в этой группе до конца теста.

Для начала посмотрим на распределение посетителей по группам А и B. 

```python
visitors_count = visitors.groupby('group').agg({'visitors':'sum'}).reset_index()
```

```python
print(f'Количество посетителей в данных visitors (A): {visitors_count[visitors_count["group"] == "A"]["visitors"].values[0]}')
print(f'Количество посетителей в данных Orders (B): {visitors_count[visitors_count["group"] == "B"]["visitors"].values[0]}')
print('')
print(f'Относительное различие количества посетителей групп А и B : {((visitors_count.loc[0,"visitors"]/visitors_count.loc[1,"visitors"])-1)*100}%')
```

Из полученного результата видно, что разница по кол-ву посетителей составила менее 1 %. Значит мы можем быть уверены, что разделение посетителей по группам в ходе эксперимента произошло корректно.

Теперь проверим, что попавший в одну из групп пользователь остаётся в этой группе до конца теста. Если пользователь видит разные версии исследуемой страницы в ходе одного исследования, неизвестно, какая именно повлияла на его решения. Значит, и результаты такого теста нельзя интерпретировать однозначно.

```python
# временные рамки теста
print(f'Количество заказов в данных Orders (A): {orders[orders["group"] == "A"]["transactionId"].nunique()}')
print(f'Количество заказов в данных Orders (B): {orders[orders["group"] == "B"]["transactionId"].nunique()}')
print('')
print(f'Количество пользователей в данных Orders (A): {orders[orders["group"] == "A"]["visitorId"].nunique()}')
print(f'Количество пользователей в данных Orders (B): {orders[orders["group"] == "B"]["visitorId"].nunique()}')


# пересечение пользователей между группами
visitors_a = set(orders[orders["group"] == "A"]["visitorId"].unique())
visitors_b = set(orders[orders["group"] == "B"]["visitorId"].unique())
visitors_ab = visitors_a.intersection(visitors_b)
print(f'Количество пересекающихся пользователей между группами A и B: {len(visitors_ab)}')
```

Проводить дальнейший анализ с такими пересекающимися пользователями не имеет смысла, так как они могут повлиять на результат проводимых тестов. ### График кумулятивной выручки по группам

В рамках эксперимента построим график кумулятивной выручки по группам. Сделаем выводы и предположения.

```python
# создаем массив уникальных пар значений дат и групп теста
datesGroups = orders[['date','group']].drop_duplicates()

# получаем агрегированные кумулятивные по дням данные о заказах 
ordersAggregated = datesGroups.apply(lambda x: orders[np.logical_and(orders['date'] <= x['date'], orders['group'] == x['group'])].agg({'date' : 'max', 'group' : 'max', 'transactionId' : 'nunique', 'visitorId' : 'nunique', 'revenue' : 'sum'}), axis=1).sort_values(by=['date','group'])

# получаем агрегированные кумулятивные по дням данные о посетителях интернет-магазина 
visitorsAggregated = datesGroups.apply(lambda x: visitors[np.logical_and(visitors['date'] <= x['date'], visitors['group'] == x['group'])].agg({'date' : 'max', 'group' : 'max', 'visitors' : 'sum'}), axis=1).sort_values(by=['date','group'])

# объединяем кумулятивные данные в одной таблице и присваиваем ее столбцам понятные названия
cumulativeData = ordersAggregated.merge(visitorsAggregated, left_on=['date', 'group'], right_on=['date', 'group'])
cumulativeData.columns = ['date', 'group', 'orders', 'buyers', 'revenue', 'visitors']

print(cumulativeData.head(5)) 
```

Построим графики кумулятивной выручки по дням и группам A/B-тестирования: 

```python
# датафрейм с кумулятивным количеством заказов и кумулятивной выручкой по дням в группе А
cumulativeA = cumulativeData[cumulativeData['group']=='A'][['date','revenue', 'orders','visitors']]

# датафрейм с кумулятивным количеством заказов и кумулятивной выручкой по дням в группе B
cumulativeB = cumulativeData[cumulativeData['group']=='B'][['date','revenue', 'orders','visitors']]

# Применяем стиль к текущему графику
plt.style.use('fivethirtyeight')

plt.figure(figsize=(15, 12))

# Строим график выручки группы А
plt.plot(cumulativeA['date'], 
         cumulativeA['revenue'], 
         label='A')

# Строим график выручки группы B
plt.plot(cumulativeB['date'], 
         cumulativeB['revenue'], 
         label='B')

plt.legend()
plt.ylabel('Кумулятивная выручка', fontsize=20)
plt.xlabel('Дата эксперимента',fontsize=20)
plt.title('График кумулятивной выручки по дням и группам A/B-тестирования',fontsize=23)
plt.show();
```

В самом начале эксперимента выручка группы В была немного ниже, чем у группы А, однако уже на шестой день эксперимента начала увеличиваться и оставалась ведущей на протяжении всего периода А/Б-теста. Выручка по группе А показывала почти равномерный рост в течение всего теста, в то время как в группе Б наблюдался резкий скачок выручки в период с 18 по 20 числа. Этот феномен может указывать на возможные всплески в числе заказов или появление очень дорогих заказов в данной выборке.

### График кумулятивного среднего чека по группам

В рамках эксперимента построим график кумулятивного среднего чека по группам. Для этого разделим кумулятивную выручку на кумулятивное число заказов. Сделаем выводы и предположения.

```python
plt.figure(figsize=(15,12))
plt.plot(cumulativeA['date'], 
         cumulativeA['revenue']/cumulativeA['orders'], 
         label='A')

plt.plot(cumulativeB['date'], 
         cumulativeB['revenue']/cumulativeB['orders'], 
         label='B')


plt.legend()
plt.ylabel('Кумулятивный средний чек', fontsize=20)
plt.xlabel('Дата эксперимента',fontsize=20)
plt.title('График кумулятивного среднего чека по дням и группам A/B-тестирования',fontsize=23)
plt.show();
```

### График относительного изменения кумулятивного среднего чека

В рамках эксперимента построим график относительного изменения кумулятивного среднего чека группы B к группе A. Сделаем выводы и предположения.

```python
plt.figure(figsize=(15,12))

# собираем данные в одном датафрейме
mergedCumulativeAB = cumulativeA.merge(cumulativeB, 
                                                   left_on='date', 
                                                   right_on='date', 
                                                   how='left', 
                                                   suffixes=['A', 'B'])

# cтроим отношение средних чеков
plt.plot(mergedCumulativeAB['date'], 
        (mergedCumulativeAB['revenueB']/mergedCumulativeAB['ordersB'])
         /(mergedCumulativeAB['revenueA']/mergedCumulativeAB['ordersA'])-1)

# добавляем ось X
plt.axhline(y=0, color='black', linestyle='--')

plt.ylabel('Относительное изменение', fontsize=20)
plt.xlabel('Дата эксперимента',fontsize=20)
plt.title('График относительного изменения кумулятивного среднего чека по дням группы B к А',fontsize=23)
plt.show();
```

### Статистическая значимость различий в среднем количестве заказов на посетителя между группами по «сырым» данным.

Сформулируем гипотезы. 
- `H0: различий в среднем количестве заказов на посетителя  между группами нет.`
- `H1: различия в среднем количестве заказов на посетителя  между группами есть.`

Уровень статистической значимости зададим alpha = 0.05  (5%)

По подготовленным данным посчитаем статистическую значимость различий в среднем количестве заказов на посетителя между группами по «сырым» данным. Сделаем выводы и предположения.

Начнём с подготовки данных. 

**Для расчёта значений статистической значимости для каждой группы теста необходимо создать таблицу в которой будут следующая информация:**

- date — дата;
- ordersPerDateA — количество заказов в выбранную дату в группе A;
- revenuePerDateA — суммарная выручка в выбранную дату в группе A;
- ordersPerDateB — количество заказов в выбранную дату в группе B;
- revenuePerDateB — суммарная выручка в выбранную дату в группе B;
- ordersCummulativeA — суммарное число заказов до выбранной даты включительно в группе A;
- revenueCummulativeA — суммарная выручка до выбранной даты включительно в группе A;
- ordersCummulativeB — суммарное количество заказов до выбранной даты включительно в группе B;
- revenueCummulativeB — суммарная выручка до выбранной даты включительно в группе B;
- visitorsPerDateA — количество пользователей в выбранную дату в группе A;
- visitorsPerDateB — количество пользователей в выбранную дату в группе B;
- visitorsCummulativeA — количество пользователей до выбранной даты включительно в группе A;
- visitorsCummulativeB — количество пользователей до выбранной даты включительно в группе B.

```python
visitorsADaily = visitors[visitors['group'] == 'A'][['date', 'visitors']]
visitorsADaily.columns = ['date', 'visitorsPerDateA']

visitorsACummulative = visitorsADaily.apply(
    lambda x: visitorsADaily[visitorsADaily['date'] <= x['date']].agg(
        {'date': 'max', 'visitorsPerDateA': 'sum'}
    ),
    axis=1,
)
visitorsACummulative.columns = ['date', 'visitorsCummulativeA']

visitorsBDaily = visitors[visitors['group'] == 'B'][['date', 'visitors']]
visitorsBDaily.columns = ['date', 'visitorsPerDateB']

visitorsBCummulative = visitorsBDaily.apply(
    lambda x: visitorsBDaily[visitorsBDaily['date'] <= x['date']].agg(
        {'date': 'max', 'visitorsPerDateB': 'sum'}
    ),
    axis=1,
)
visitorsBCummulative.columns = ['date', 'visitorsCummulativeB']

ordersADaily = (
    orders[orders['group'] == 'A'][['date', 'transactionId', 'visitorId', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'transactionId': pd.Series.nunique, 'revenue': 'sum'})
)
ordersADaily.columns = ['date', 'ordersPerDateA', 'revenuePerDateA']

ordersACummulative = ordersADaily.apply(
    lambda x: ordersADaily[ordersADaily['date'] <= x['date']].agg(
        {'date': 'max', 'ordersPerDateA': 'sum', 'revenuePerDateA': 'sum'}
    ),
    axis=1,
).sort_values(by=['date'])
ordersACummulative.columns = [
    'date',
    'ordersCummulativeA',
    'revenueCummulativeA',
]

ordersBDaily = (
    orders[orders['group'] == 'B'][['date', 'transactionId', 'visitorId', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'transactionId': pd.Series.nunique, 'revenue': 'sum'})
)
ordersBDaily.columns = ['date', 'ordersPerDateB', 'revenuePerDateB']

ordersBCummulative = ordersBDaily.apply(
    lambda x: ordersBDaily[ordersBDaily['date'] <= x['date']].agg(
        {'date': 'max', 'ordersPerDateB': 'sum', 'revenuePerDateB': 'sum'}
    ),
    axis=1,
).sort_values(by=['date'])
ordersBCummulative.columns = [
    'date',
    'ordersCummulativeB',
    'revenueCummulativeB',
]

data = (
    ordersADaily.merge(
        ordersBDaily, left_on='date', right_on='date', how='left'
    )
    .merge(ordersACummulative, left_on='date', right_on='date', how='left')
    .merge(ordersBCummulative, left_on='date', right_on='date', how='left')
    .merge(visitorsADaily, left_on='date', right_on='date', how='left')
    .merge(visitorsBDaily, left_on='date', right_on='date', how='left')
    .merge(visitorsACummulative, left_on='date', right_on='date', how='left')
    .merge(visitorsBCummulative, left_on='date', right_on='date', how='left')
)

print(data.head(5)) 
```

Создадим переменные `ordersByUsersA` и `ordersByUsersB` со столбцами ['userId', 'orders']. В них для пользователей, которые заказывали хотя бы 1 раз, укажем число совершённых заказов. 

```python
# Количество заказов для каждого уникального пользователя в группе А
ordersByUsersA = (
    orders[orders['group'] == 'A']
    .groupby('visitorId', as_index=False)
    .agg({'transactionId': pd.Series.nunique})
)
ordersByUsersA.columns = ['userId', 'orders']

# Количество заказов для каждого уникального пользователя в группе В
ordersByUsersB = (
    orders[orders['group'] == 'B']
    .groupby('visitorId', as_index=False)
    .agg({'transactionId': pd.Series.nunique})
)
ordersByUsersB.columns = ['userId', 'orders'] 
```

Объявим переменные `sampleA` и `sampleB`, в которых пользователям из разных групп будет соответствовать количество заказов. Тем, кто ничего не заказал, будут соответствовать нули. Это нужно, чтобы подготовить выборки к проверке критерием Манна-Уитни.

Переменная sampleA должна состоять из двух частей:
Список с количеством заказов для каждого из пользователей: ordersByUsersA['orders'].
Нули для пользователей, которые ничего не заказывали. Их количество равно разнице между суммой посетителей и количеством записей о заказах: data['visitorsPerDateA'].sum() - len(ordersByUsersA['orders']). Создадим объект pd.Series нужной длины.

Дальше передадим аргумент, сообщающий, что объекты Series нужно объединить по строкам (то есть записать подряд): pd.concat([...], axis=0). То же делаем для sampleB и получаем:

```python
# Таблица с количеством покупок для каждого клиента и если клиент ничего не купил то значение покупок равно 0 [Группа А]
sampleA = pd.concat(
    [
        ordersByUsersA['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)


# Таблица с количеством покупок для каждого клиента и если клиент ничего не купил то значение покупок равно 0 [Группа B]
sampleB = pd.concat(
    [
        ordersByUsersB['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateB'].sum() - len(ordersByUsersB['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)


 
```

1. Применим критерий и отформатируем p-value, округлив его до трёх знаков после запятой.Задействуем статистический критерий Манна-Уитни к полученным выборкам
2. Выведем относительный прирост среднего числа заказов группы B: среднее число заказов группы B / среднее число заказов группы A - 1.  Округлим до трёх знаков после запятой.

```python
alpha = 0.05 # задаём уровень статистической значимости в 5%

results = stats.mannwhitneyu(sampleA, sampleB)[1]
```

```python
print('p-значение:{0:.3f}'.format(results))

if results < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

```python
print("Относительное изменение в среднем числе заказов группы B к A: {0:.3f}".format(sampleB.mean() / sampleA.mean() - 1))
```

- По результатам статистического теста мы можем сделать вывод, что **имеются различия в среднем числе заказов групп A и B, так как p-value = 0.011(1.1%), что меньше уровня статистической значимости в 5%. Соответсвенно есть основания отвергнуть нулевую гипотезу о равенстве среднего кол-ва заказов на посетителя между группами.**
- По результатам относительного изменения видно превосходство в среднем числе заказов группы B на 16,0%

### Cтатистическая значимость различий в среднем чеке заказа между группами по «сырым» данным. 

Посчитаем статистическую значимость различий в среднем чеке заказа между группами по «сырым» данным. Сделаем выводы и предположения.

Сформулируем гипотезы. 
- `H0: различий в среднем чеке заказа между группами нет.`
- `H1: различия в среднем чеке заказа между группами есть.`

Уровень статистической значимости зададим alpha = 0.05  (5%)

1. Чтобы рассчитать статистическую значимость различий в среднем чеке, передадим критерию mannwhitneyu() данные о выручке с заказов. Применим критерий и отформатируем p-value, округлив его до трёх знаков после запятой.Задействуем статистический критерий Манна-Уитни к полученным выборкам
2. Выведем относительный прирост среднего чека заказов группы B: средний чек заказов группы B / средний чек заказов группы A - 1. Округлим до трёх знаков после запятой.

- Однако по результатам относительного изменения видно превосходство в среднем чеке заказов группы B на 28,7%

### Статистическая значимость различий в среднем количестве заказов на посетителя между группами по «очищенным» данным.

1. Ранее мы уже выяснили, что  не более 5% пользователей оформляли больше одного заказа. Разумно выбрать пользователей с одним заказом за верхнюю границу числа заказов, и отсеять аномальных пользователей по ней. 

2. Также не более 5% заказов имеют цену, превышающую 26785 у.е. Таким образом, разумным подходом будет отобрать заказы с ценой до 26785 у.е. 

Очистим данные от аномальных пользователей с превышающими данные лимиты значениями.

Узнаем, сколько всего аномальных пользователей атрибутом shape

```python
usersWithManyOrders = pd.concat(
    [
        ordersByUsersA[ordersByUsersA['orders'] > np.percentile(ordersByUsers['orders'],95)]['userId'],
        ordersByUsersB[ordersByUsersB['orders'] > np.percentile(ordersByUsers['orders'],95)]['userId'],
    ],
    axis=0,
)
usersWithExpensiveOrders = orders[orders['revenue'] > np.percentile(orders['revenue'],95)]['visitorId']
abnormalUsers = (
    pd.concat([usersWithManyOrders, usersWithExpensiveOrders], axis=0)
    .drop_duplicates()
    .sort_values()
)
print(abnormalUsers.head(5))
print(abnormalUsers.shape[0])
```

У нас имеется 86 аномальных пользователей. Подготовим выборки количества заказов по пользователям по группам теста:

```python
sampleAFiltered = pd.concat(
    [
        ordersByUsersA[
            np.logical_not(ordersByUsersA['userId'].isin(abnormalUsers))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)

sampleBFiltered = pd.concat(
    [
        ordersByUsersB[
            np.logical_not(ordersByUsersB['userId'].isin(abnormalUsers))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateB'].sum() - len(ordersByUsersB['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)
```

Узнаем, как их действия повлияли на результаты теста. Посчитаем статистическую значимость различий в среднем количестве заказов между группами теста по очищенным данным. 

Сформулируем гипотезы. 

- `H0: различий в среднем количестве заказов на посетителя между группами нет.`
- `H1: различия в среднем количестве заказов на посетителя между группами есть.`

Уровень статистической значимости зададим alpha = 0.05  (5%)

1. Применим критерий и отформатируем p-value, округлив его до трёх знаков после запятой.Задействуем статистический критерий Манна-Уитни к полученным выборкам
2. Выведем относительный прирост среднего числа заказов группы B: среднее число заказов группы B / среднее число заказов группы A - 1.  Округлим до трёх знаков после запятой.

```python
alpha = 0.05 # задаём уровень статистической значимости в 5%

results = stats.mannwhitneyu(sampleAFiltered, sampleBFiltered)[1]
```

```python
print('p-значение:{0:.3f}'.format(results))

if results < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

```python
print("Относительное изменение в среднем числе заказов группы B к A: {0:.3f}".format(sampleBFiltered.mean() / sampleAFiltered.mean() - 1))
```

- По результатам статистического теста мы можем сделать вывод, что **имеются различия в среднем числе заказов групп A и B, так как p-value = 0.016(1.6%), что меньше уровня статистической значимости в 5%. Соответсвенно есть основания отвергнуть нулевую гипотезу о равенстве среднего кол-ва заказов на посетителя между группами.**

Мы можем сделать вывод, что результаты по среднему количеству заказов практически не изменились.Сырые данные не повлияли на работу теста.Группа В всё также лидирует по относительному изменению.

### Статистическая значимость различий в среднем чеке заказа между группами по «очищенным» данным. 

Посчитаем статистическую значимость различий в среднем чеке заказа между группами по «очищенным» данным. Сделаем выводы и предположения.

Сформулируем гипотезы. 
- `H0: различий в среднем чеке заказа между группами нет.`
- `H1: различия в среднем чеке заказа между группами есть.`

Уровень статистической значимости зададим alpha = 0.05  (5%)

1. Чтобы рассчитать статистическую значимость различий в среднем чеке, передадим критерию mannwhitneyu() данные о выручке с заказов. Применим критерий и отформатируем p-value, округлив его до трёх знаков после запятой.Задействуем статистический критерий Манна-Уитни к полученным выборкам
2. Выведем относительный прирост среднего чека заказов группы B: средний чек заказов группы B / средний чек заказов группы A - 1. Округлим до трёх знаков после запятой.

```python
alpha = 0.05 # задаём уровень статистической значимости в 5%

results = stats.mannwhitneyu(
           orders[np.logical_and(
        orders['group']=='A',
        np.logical_not(orders['visitorId'].isin(abnormalUsers)))]['revenue'],
    orders[np.logical_and(
        orders['group']=='B',
        np.logical_not(orders['visitorId'].isin(abnormalUsers)))]['revenue'])[1]
```

```python
print('p-значение:{0:.3f}'.format(results))

if results < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

```python
print("Относительное изменение в среднем чеке заказов группы B к A: {0:.3f}".format(orders[np.logical_and(orders['group']=='B',
                                                                                                          np.logical_not(orders['visitorId'].isin(abnormalUsers)))]['revenue'].mean()/
    orders[np.logical_and(
        orders['group']=='A',
        np.logical_not(orders['visitorId'].isin(abnormalUsers)))]['revenue'].mean() - 1))
```

- По результатам статистического теста мы можем сделать вывод, что **нет оснований полагать о различии в среднем чеке заказов между группами A и B, так как p-value = 0.727(72.7%), что больше уровня статистической значимости в 5%. Соответсвенно нет причин отвергнуть нулевую гипотезу.**
- Однако по результатам относительного изменения видно небольшое отставание в среднем чеке заказов группы B на 3,4%

Мы можем сделать вывод, что значение статистической значимости для среднего чека между группами по "очищенным" данным фактически аналогично значению, полученному на основе "сырых" данных. Однако, после удаления выбросов, стало видно, что средний чек в группе B на 3,4% ниже, чем в группе A. Этот показатель существенно отличается от результата, полученного на исходных данных, где превосходство в среднем чеке заказов группы B было 28,7% . В данном случае очистка данных от аномалий имела значительное влияние на итоговый результат и позволила более объективно оценить статистику.
