Узнаем численность таких пользователей, сначала  для группы А нашего теста, потом для Б.

Теперь найдём кол-во пересеченных пользователей группы Б (конкурирующий тест) и пользователей группы А и Б по отдельности (наш тест).

```python
recommender_system_test_A = set(participants_EU[(participants_EU['ab_test'] == 'recommender_system_test') &
                                                (participants_EU['group'] == 'A')]['user_id'].unique())
interface_eu_test_A = set(participants_EU[(participants_EU['ab_test'] == 'interface_eu_test') &
                                          (participants_EU['group'] == 'B')]['user_id'].unique())

print(f'Пересечение пользователей группы Б (конкурирующий тест) и пользователей группы А (наш тест): {len(recommender_system_test_A.intersection(interface_eu_test_A))} ч.')
```

```python
recommender_system_test_B = set(participants_EU[(participants_EU['ab_test'] == 'recommender_system_test') &
                                                (participants_EU['group'] == 'B')]['user_id'].unique())
interface_eu_test_B = set(participants_EU[(participants_EU['ab_test'] == 'interface_eu_test') &
                                          (participants_EU['group'] == 'B')]['user_id'].unique())

print(f'Пересечение пользователей группы Б (конкурирующий тест) и пользователей группы Б (наш тест): {len(recommender_system_test_B.intersection(interface_eu_test_B))} ч.')
```

Группа Б конкурирующего теста могла повлиять на пересечённых пользователей, так как пользователям тестируемой группы Б всегда показываются новые фичи продукта. Логично было бы удалить этих пользователей, но мы должны узнать как они распределены между группами нашего теста `recommender_system_test`. Если доли этих пересечённых пользователей распределены равномерно между группами А и Б нашего теста, то их можно оставить.

Осуществлять проверку на возможную стат. значимость различий долей между группами А и B нашего теста мы будем с помощью проверки гипотезы о равенстве долей(z тест). Мы будем сравнивать пропорции(Кол-во пересечённых пользователей группы Б (конкурирующий тест) и пользователей группы А (наш тест) к общей численности пользователей нашего теста в группе А с пропорцией (Кол-во пересечённых пользователей группы Б (конкурирующий тест) и пользователей группы Б (наш тест) к общей численности пользователей нашего теста в группе Б). Сами группы будут разделены на А и Б.Разница между пропорциями, наблюдаемыми в группах, будет нашей статистикой. Всего будет проведён 1 тест.

Для проведения теста сформулируем гипотезы:

- Н0 : доли групп  равны
- Н1 : доли групп  равны  неравны

Воспользуемся готовым тестом `proportions_ztest` для проверки гипотезы. Проверим, будет ли отличие между группами А и B статистически значимым. Зададим пороговый уровень статистической значимости alpha = 0.05 (5% вероятности), что результат получен случайно и не соответствует действительности,вероятность ошибки первого рода.

```python
alpha = 0.05 # задаём уровень статистической значимости в 5%

crossed_users_by_groups = [len(recommender_system_test_A.intersection(interface_eu_test_A)), len(recommender_system_test_B.intersection(interface_eu_test_B))]
count_user_groups = [user_distribution.loc['2020-12-21','cumulativeA'], user_distribution.loc['2020-12-21','cumulativeB']]
results = proportions_ztest(crossed_users_by_groups, count_user_groups)[1]
print('p-значение:{0:.3f}'.format(results))

if results < alpha:
    print('Отвергаем нулевую гипотезу, доли групп равны неравны')
else:
    print('Не получилось отвергнуть нулевую гипотезу,доли групп равны')
```

Статистический тест подтвердил равномерное распределение пересекающихся пользователей по группам нашего теста `recommender_system_test`. Негативное влияние пересекающихся пользователей будет оказываться в равной степени по группам. Примем решение и оставим пересекающихся пользователей. Тем не менее для будущих тестов, необходимо исключить влияние такого рода ошибок, так как это напрямую влияет на результат проводимого тестирования.

### Проверка на пересечение пользователей между группами

### Маркетинговые события и пользовательская активность

Необходимо проверить пересечение маркетинговых событий с периодом проведения теста `recommender_system_test`. Оценим воздействие маркетинговых событий на динамику количества событий пользователей.

Настроим автоматическую проверку, выдающую список событий, пересекающихся с нашим тестом.

```python
# Задаём даты фактического проведения теста
start_test = pd.to_datetime(datetime(2020, 12, 7))
end_test = pd.to_datetime(datetime(2020, 12, 30))
# Задаём целевой регион
region = 'EU'
```

Обозначим все все виды событий, которые могут пересечься с нашим периодом:

- Начались до периода, закончились во время
- Начались до периода, закончились после
- Начались во время периода, закончились во время периода
- Начались во время периода, закончились после

Исходя из представленного алгоритма, создадим функцию `filter_events`, которая на вход, в качестве параметров функции будет принимать: 1)датафрейм с информацией по маркетинговым событиям 2)Дату начала проведения теста теста 3) Дату конца проведения теста теста.4) Целевой регион. По результатам применения функции будут оставлены только те маркетинговые события, которые соответствуют указанному диапазону проводимого теста.

```python
def filter_events(df, start_period, end_period,region):
    # пустой список filtered_events, который будет содержать отфильтрованные события
    filtered_events = []
    # Итерация по строкам DataFrame df с использованием iterrows()
    for index, row in df.iterrows():
        # Извлечение информации о времени начала (event_start), времени завершения (event_end) и
        # регионах (event_regions) для каждого события
        event_start = pd.to_datetime(row['start_dt'])
        event_end = pd.to_datetime(row['finish_dt'])
        event_regions = [r.strip() for r in row['regions'].split(',')]

        # Проверка условий
        if (event_start < start_period and event_end >= start_period and event_end <= end_period) or \
           (event_start < start_period and event_end > end_period) or \
           (event_start >= start_period and event_end <= end_period) or \
           (event_start >= start_period and event_start <= end_period and event_end > end_period) and \
           (region in event_regions):

            # Добавление информации о событии в filtered_events
            filtered_events.append((row['name'], row['regions'], row['start_dt'], row['finish_dt']))

    filtered_df = pd.DataFrame(filtered_events, columns=df.columns)
    return filtered_df
```

### Продуктовая воронка

Построим простые продуктовые воронки для двух групп теста с учетом логической последовательности совершения событий. Изучим изменение конверсии в продуктовой воронке тестовой группы, по сравнению с контрольной.

Для начала необходимо понять какие этапы событий существуют в нашей таблице

```python
filtred_data['event_name'].value_counts()
```

Ранее, в ходе проведённого анализа мы выяснили,что большая доля пользователей не совершала каких-либо событий в обеих группах. Мы приняли решение оставить этих пользователей для дальнейших исследований.Необходимо учесть их для формирования продуктовой воронки. Так как они не совершали никаких событий, но числятся в таблице с данными по регистрациям пользователей, будем считать,что это пользователи, которые совершили только регистрацию. Соответственно регистрация будет являться первым этапом воронки. Целевое действие purchase (покупка) последнее в воронке.

На основании выше сказанного сформируем логическую последовательность совершения событий:
- ↓ registration
- ↓ login
- ↓ product_page
- ↓ product_cart
-   purchase

Посчитаем, сколько уникальных пользователей совершали каждое из этих событий в разрезе групп.

```python
user_events_distribution = filtred_data.pivot_table(index='event_name',
                                                    columns='group',
                                                    values='user_id',
                                                    aggfunc={'user_id':'nunique'}).reset_index()

#  убираем название оси для первого столбца в DataFrame.
user_events_distribution = user_events_distribution.rename_axis(None, axis=1)
user_events_distribution
```

Теперь посчитаем общее кол-во уникальных пользователей в разрезе групп.Полученные значения будут отражать кол-во зарегистрированных пользователей, по группам.

```python
registration_distribution = filtred_data.pivot_table(index='group',
                                                     values='user_id',
                                                     aggfunc={'user_id':'nunique'}).reset_index()
registration_distribution
```

Полученную таблицу необходимо привести к формату `user_events_distribution` для дальнейшего присоединения к ней. Для этого нужно транспонировать (перевернуть) таблицу и переименовать колонки в А и B.

```python
# Транспонируем таблицу
registration_distribution = registration_distribution.transpose()

registration_distribution = registration_distribution.reset_index()

registration_distribution
```

```python
# Удаляем строку с индексом 0
registration_distribution.drop(index=0,inplace=True)
registration_distribution.columns = ['event_name','A','B']
registration_distribution.loc[1,'event_name'] = 'registration'
registration_distribution
```

Теперь соеденим полученные 2-е таблицы в одну. Назовём её   `purchase_funnel`

```python
purchase_funnel = pd.concat([user_events_distribution,registration_distribution])

# Cоздаём переменную со списком индексов
index = [0,1,2,3,4]

# Установка индекса для таблицы purchase_funnel
purchase_funnel = purchase_funnel.set_index(pd.Index(index))

purchase_funnel
```

Поменяем местами строки в соответствиии с логической последовательностью совершения событий пользователями

```python
# Вручную меняем местами строки с помощью индексов

purchase_funnel = purchase_funnel.reindex([4,0,2,1,3])
purchase_funnel = purchase_funnel.reset_index(drop=True)
purchase_funnel
```

Перед расчётом конверсии изменим тип данных по столбцам 'A' и 'B' на целочисленный.

```python
purchase_funnel['A'] = pd.to_numeric(purchase_funnel['A'])
purchase_funnel['B'] = pd.to_numeric(purchase_funnel['B'])
```

Произведём расчёт конверсии в каждый шаг от начального шага воронки (registration) по каждой группе, полученные значения укажем в % округлённые до 2-ух знаков.

```python
purchase_funnel['conversion_A_%'] = round((purchase_funnel['A']/purchase_funnel.loc[0,'A'])*100,2)
purchase_funnel['conversion_B_%'] = round((purchase_funnel['B']/purchase_funnel.loc[0,'B'])*100,2)
```

Дополним нашу таблицу столбцом `conversion_difference` с расчётом относительного различия конверсии группы B к A по каждому событию

```python
purchase_funnel['conversion_difference_%'] = round((1-(purchase_funnel['conversion_B_%']/purchase_funnel['conversion_A_%']))*100,2)
purchase_funnel
```

## Проверка гипотезы

Цель проводимого А/Б тестирования - это проверка на улучшение показателей конверсии связанных с внедрением улучшенной рекомендательной системы в тестируемой группе по каждому этапу продуктовой воронки. Ожидаемый эффект: за 14 дней с момента регистрации в системе пользователи покажут улучшение каждой метрики не менее, чем на 5 процентных пунктов. К сожалению, мы не можем быть уверены в правдоподобности результатов проводимого тестирования. Проведённая оценка на предмет корректности проведения теста и его соответствия требованиям техническому заданию показала ряд отклонений.

Сформируем гипотезу: **Проверка гипотезы о различии конверсии в событие от общего числа зарегистрированных пользователей между группами А и Б**. Результат проверки данной гипотезы поможет менеджерам проработать дальнейшие мероприятия для корректировки продуктового вектора.

Для проверки гипотезы мы будем использовать созданную ранее таблицу `purchase_funnel` с информацией по кол-ву пользователей на каждом этапе воронки разделённую на группы А и Б. Важно понять, отличается ли конверсия по каждому этапу продуктовой воронки между этими группами.

Осуществлять проверку на возможную стат. значимость различий между группами А и B мы будем с помощью проверки гипотезы о равенстве долей(z тест). Мы будем сравнивать пропорции(конверсию в: login,product_page,product_cart,purchase) из registration двух групп. Сами группы будут разделены на А и Б. Разница между пропорциями, наблюдаемыми в группах, будет нашей статистикой.  Всего будет проведено 4 теста.

Для проведения 4-ёх тестов сформулируем гипотезы:

- Н0 : конверсии групп  равны
- Н1 : конверсии групп  неравны

Для проверки гипотезы воспользуемся готовым тестом `proportions_ztest`. Проверим, будет ли отличие между группами А и B статистически значимым. Зададим пороговый уровень статистической значимости alpha = 0.05 (5% вероятности), что результат получен случайно и не соответствует действительности,вероятность ошибки первого рода.

Мы уже знаем, что будет проведено 4 теста, если оставим  пороговый уровень статистической значимости alpha = 0.05 (5% вероятности), то произойдёт увеличечение ошибки первого рода,так называемая групповая вероятность ошибки первого рода, или FWER. Несколько сравнений, проводимых на одних и тех же данных — это множественный тест.

Найдём вероятность того, что хотя бы в одном из четырёх попарных сравнений будет зафиксирован ложнопозитивный результат. Для этого применим формулу `1-(1- alpha)**n`, где n - кол-во тестов.Если полученная вероятность ошибки будет меньше уровня значимости - выбранный уровень значимости оптимальный, если полученная вероятность ошибки будет больше уровня значимости - есть необходимость в снижении вероятности ложнопозитивного результата.

```python
alhpa = 0.05

FWER = 1-(1-alhpa)**16

if FWER < alhpa:
    print(f'Групповая вероятность ошибки первого рода {FWER}')
    print(f'Выбранный уровень стат. значимости {alhpa} оптимальный')
else:
    print(f'Групповая вероятность ошибки первого рода {FWER}')
    print(f'Выбранный уровень стат. значимости {alhpa} не оптимальный,есть необходимость в снижении вероятности ложнопозитивного результата')
```

Полученный результат сообщает нам о том, что есть необходимость в снижении вероятности ложнопозитивного результата.Чтобы снизить вероятность ложнопозитивного результата при множественном тестировании гипотез, применяют разные методы корректировки уровня значимости для уменьшения FWER.Мы воспользуемся методом Бонферрони (поправка Бонферрони).Самая грубая и распространённая поправка к требуемым уровням значимости. Уровни значимости в каждом из m сравнений в m раз меньше, чем уровень значимости, требуемый при единственном сравнении. Проще говоря, делят уровень значимости ɑ на число тестов:


```python
bonf_alpha = alhpa/4
bonf_alpha
```

Мы нашли необходимый уровень стат. значимости для выполнения 4-ёх тестов.

Для удобства применения стат.теста модернизируем нашу таблицу `purchase_funnel`. Сделаем столбец event_name индексным <a id="таблица"></a>

Создадим универсальную функцию для проверки гипотезы о равенстве долей `z_test`,где параметрами будут являться:
- event_name(Название события)
- alpha (уровень стат значимости, вероятность ошибки первого рода)

Возвращаемыми значениями будут:
- A (обозначение 1-ой группы )
- B (обозначение 2-ой группы)
- event_name (Название события)
- user_count_event_group_A (число пользователей первой группы)
- user_count_event_group_B (число пользователей второй группе)
- results (p-уровень значимости)
- отвергаем H0, конверсии групп неравны
- не отвергаем H0, конверсии групп равны

```python
def z_test(event_name, alpha):

    #выбор ячейки таблицы с уникальным кол-вом пользователей, по конкретной группе и событию
    user_count_event_group_A = purchase_funnel.loc[event_name, 'A']
    user_count_event_group_B = purchase_funnel.loc[event_name, 'B']

    #выбор ячейки таблицы с уникальным кол-вом пользователей зарегистрированных в группе А
    user_count_group_A = purchase_funnel.loc['registration', 'A']

    #выбор ячейки таблицы с уникальным кол-вом пользователей зарегистрированных в группе Б
    user_count_group_B = purchase_funnel.loc['registration', 'B']


    event = user_count_event_group_A,user_count_event_group_B
    registration = user_count_group_A,user_count_group_B

    # Применяем готовый стат. критерий
    results = proportions_ztest(event, registration)[1]


    # выводим полученный p-уровень значимости
    print('p-значение:{0:.3f}'.format(results))

    # задаём условную конструкцию и ветвим полученный результат,возвращаем полученные значения.
    if results < alpha:
        return ['A','B',event_name,user_count_event_group_A,user_count_event_group_B,results, 'отвергаем H0, конверсии групп неравны']
    else:
        return ['A','B',event_name,user_count_event_group_A,user_count_event_group_B,results, 'не отвергаем H0, конверсии групп равны']
```

Для удобства отображения результата выполнения функции `z_test` по всем событиям, создадим функцию `result_z_test` с помощью которой мы сможем вывести результаты работы функции `z_test` в виде таблицы,где параметром будут являться:

- alpha (уровень стат значимости, вероятность ошибки первого рода)



Возвращаемыми значениями будут:
- group1 (обозначение А группы)
- group2 (обозначение Б группы)
- event_name (Название события)
- user_count_event_group_A (число пользователей А группы)
- user_count_event_group_B (число пользователей Б группы)
- p_value (p-уровень значимости)
- result test (отвергаем H0, конверсии групп неравны/не отвергаем H0, конверсии групп равны)

```python
def result_z_test(alpha):
    result = []
    events = ['login', 'product_page', 'product_cart', 'purchase']

    # Уровень статистической значимости
    alpha = alpha

    # цикл для сбора результатов тестов по событиям и группам
    for event_name in events:
        result.append(z_test(event_name, alpha))  # Используем z_test для проведения теста

    result_z_test = pd.DataFrame(result, columns=['group_1', 'group_2', 'event_name','user_count_event_group_A','user_count_event_group_B', 'p_value','result test'])
    return result_z_test
```
