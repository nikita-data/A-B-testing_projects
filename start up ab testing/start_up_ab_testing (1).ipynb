### A/A-тест

Перед тем как начать A/B-тест, надо убедиться, что:
- на результаты не влияют аномалии и выбросы в генеральной совокупности
- инструмент «деления» трафика работает безошибочно
- данные отправляются в системы аналитики корректно
- попавший в одну из групп посетитель остаётся в этой группе до конца теста

Для этого проводят A/A-тест

Проверку на корректность деления трафика и сохранением за пользователем одной группы мы уже проверяли ранее
[распределение пользователей](#распределение) ,мы знаем, что разделение пользователей по группам в ходе эксперимента произошло корректно и пересекающихся пользователей между группами 246,247,248 нет.


Далее  мы проверим, находят ли статистические критерии разницу между выборками 246 и 247. Если  между значениями A и A будут существенные различия, это поможет обнаружить факторы, которые привели к искажению результатов.

Осуществлять проверку на возможную стат. значимость различий между выборками 246 и 247 мы будем с помощью `проверки гипотезы о равенстве долей(z тест)`.
Мы будем сравнивать пропорции двух генеральных совокупностей по выборкам из них по каждому событию. Пропорциями будет конверсия двух выборок. Сами выборки будут разделены на 246 и 247 контрольные группы.Разница между пропорциями, наблюдаемыми на выборках, будет нашей статистикой. Всего будет проведено 4-е теста, ровно столько же событий.

Для проведения 4-ов тестов сформулируем гипотезы:

- Н0 : конверсии групп  равны
- Н1 : конверсии групп  неравны

Для удобства применения стат. теста и расчёта конверсии, нам необходимо создать единую сводную таблицу сгруппированную по номеру группы,где столбцами будут названия события, а значениями кол-во уникальных пользователей.

```python
group_table = pd.concat([(data_filtered
                  .pivot_table(index = 'exp_id',
                               values = 'device_id_hash',
                               aggfunc = 'nunique')
                  .rename(columns={"nunique": "unique_user_count"}))
                  ,(data_filtered
                  .pivot_table(index ='exp_id',
                               columns='event_name',
                               values='device_id_hash',
                               aggfunc='nunique'))]
                 , axis=1)

group_table = group_table.drop(columns = 'Tutorial')
group_table.columns= ['unique_user_count','Cart_Screen_Appear','Main_Screen_Appear','Offers_Screen_Appear','Payment_Screen_Successful']

#Выстраиваем порядок столбцов
group_table = group_table[['unique_user_count','Main_Screen_Appear','Offers_Screen_Appear','Cart_Screen_Appear','Payment_Screen_Successful']]
group_table
```

Создадим универсальную функцию для проверки гипотезы о равенстве долей `z_test`,где параметрами будут являться:
- event_name(Название события)
- alpha (уровень стат значимости, вероятность ошибки первого рода)
- group1 (первая группа)
- group2 (вторая группа)

Возвращаемыми значениями будут:
- group1 (номер первой группы)
- group2 (номер второй группы)
- event_name (Название события)
- successes_1 (число пользователей первой группы)
- successes_2 (число пользователей второй группы)
- p_value (p-уровень значимости)
- p1 (доля пользователей 1-ой группы, которые хоть раз совершали событие в %)
- p2 (доля пользователей 2-ой группы, которые хоть раз совершали событие в %)
- p2/p1 - 1 (относительная разница между долями двух групп в %)
- отвергаем H0, конверсии групп неравны
- не отвергаем H0, конверсии групп равны

```python
def z_test(event_name, alpha, group1, group2):

    #выбор ячейки таблицы с уникальным кол-вом пользователей, по конкретной группе и событию
    successes_1 = group_table.loc[group1, event_name]
    successes_2 = group_table.loc[group2, event_name]

    # выбор ячейки таблицы с уникальным кол-вом пользователей во всей выбранной группе 1
    trials_1 = group_table.loc[group1, 'unique_user_count']

    # выбор ячейки таблицы с уникальным кол-вом пользователей во всей выбранной группе 2
    trials_2 = group_table.loc[group2, 'unique_user_count']

    # пропорция успехов в первой групп:
    p1 = successes_1/trials_1

    # пропорция успехов во второй группе:
    p2 = successes_2/trials_2

    # пропорция успехов в комбинированном датасете:
    p_combined = (successes_1 + successes_2) / (trials_1 + trials_2)

    # разница пропорций в датасетах
    difference = p1 - p2

    # считаем статистику в ст.отклонениях стандартного нормального распределения
    z_value = difference / math.sqrt(p_combined * (1 - p_combined) * (1 / trials_1 + 1 / trials_2))

    # задаем стандартное нормальное распределение (среднее 0, ст.отклонение 1)
    distr = st.norm(0, 1)

    p_value = (1 - distr.cdf(abs(z_value))) * 2

    # выводим полученный p-уровень значимости
    print('p-значение: ', round(p_value,2))

    # задаём условную конструкцию и ветвим полученный результат,возвращаем полученные значения.
    if p_value < alpha:
        return [group1, group2, event_name,successes_1,successes_2, round(p_value,2), round(p1*100,2), round(p2*100,2), round((p2/p1 - 1)*100,2), 'отвергаем H0, конверсии групп неравны']
    else:
        return [group1, group2, event_name,successes_1,successes_2, round(p_value,2), round(p1*100,2), round(p2*100,2), round((p2/p1 - 1)*100,2), 'не отвергаем H0, конверсии групп равны']
```

Узнаем результат работы функции `z_test` на примере самого популярного события `Main_Screen_Appear`. Проверим, будет ли отличие между группами 246 и 247 статистически значимым.В аргумент функции передадим alpha = 0.01 (1% вероятности), что результат получен случайно и не соответствует действительности,вероятность ошибки первого рода.

```python
result = z_test('Main_Screen_Appear', 0.01,246, 247)
result
```

Функция отработала. Из полученных результатов мы получили p-значение:  0.76, что больше заданного уровня стат. значимости, соответственно H0 не отвергаем. Конверсии групп 246 и 247 равны. В первой группе(246) по заданному событию численность пользователей составила 2450, по второй(247) 2476.Доля пользователей 1-ой группы(246), которые хоть раз совершали выбранное событие  98.63%, доля пользователей 2-ой группы(247) 98.53%.

Проведём тесты для всех событий. Полученные результаты сохраним в новый датафрейм  result_z_test_246_247.
Но перед этим создадим функцию `result_z_test` с помощью которой мы сможем вывести результаты работы функции `z_test` в виде таблицы,где параметрами будут являться:

- group1 (первая группа)
- group2 (вторая группа)
- alpha (уровень стат значимости, вероятность ошибки первого рода)



Возвращаемыми значениями будут:

- user_count_group_1 (число пользователей первой группы)
- user_count_group_2 (число пользователей второй группы)
- p_value (p-уровень значимости)
- conv_group1% (доля пользователей 1-ой группы, которые хоть раз совершали событие в %)
- conv_group2% (доля пользователей 2-ой группы, которые хоть раз совершали событие в %)
- diff_conv% (относительная разница между долями двух групп в %)
- result test (отвергаем H0, конверсии групп неравны/не отвергаем H0, конверсии групп равны)

```python
def result_z_test(group1, group2, alpha):
    result = []
    events = ['Main_Screen_Appear', 'Offers_Screen_Appear', 'Cart_Screen_Appear', 'Payment_Screen_Successful']

    # Уровень статистической значимости
    alpha = alpha

    # Двойной цикл для сбора результатов тестов по событиям и группам
    for event in events:
        result.append(z_test(event, alpha, group1, group2))  # Используем z_test для проведения теста

    result_z_test = pd.DataFrame(result, columns=['group_1', 'group_2', 'event','user_count_group_1','user_count_group_2', 'p_value', 'conv_group1%', 'conv_group2%','diff_conv%','result test'])
    return result_z_test  # Возвращаем результат
```

### Корректировка уровня статистической значимости

По всем тестам нами был выбран уровень статистической значимости alpha = 0.01. Нами были проведены сравнения 3 групп + 1 одна объединённая группа,  по 4 разным событиям.Всего 16 тестов. Увеличение количества групп способствут увеличечению ошибки первого рода,так называемая групповая вероятность ошибки первого рода, или FWER. Несколько сравнений, проводимых на одних и тех же данных — это множественный тест.

Найдём вероятность того, что хотя бы в одном из четырёх попарных сравнений будет зафиксирован ложнопозитивный результат. Для этого применим формулу `1-(1- alpha)**n`, где n - кол-во тестов.Если полученная вероятность ошибки будет меньше уровня значимости - выбранный уровень значимости оптимальный, если полученная вероятность ошибки будет больше уровня значимости - есть необходимость в снижении вероятности ложнопозитивного результата.

```python
alhpa = 0.01

FWER = 1-(1-alhpa)**16

if FWER < alhpa:
    print(f'Групповая вероятность ошибки первого рода {FWER}')
    print(f'Выбранный уровень стат. значимости {alhpa} оптимальный')
else:
    print(f'Групповая вероятность ошибки первого рода {FWER}')
    print(f'Выбранный уровень стат. значимости {alhpa} не оптимальный,есть необходимость в снижении вероятности ложнопозитивного результата')
```

Полученный результат сообщает нам о том, что есть необходимость в снижении вероятности ложнопозитивного результата.Чтобы снизить вероятность ложнопозитивного результата при множественном тестировании гипотез, применяют разные методы корректировки уровня значимости для уменьшения FWER.Мы воспользуемся методом Бонферрони (поправка Бонферрони).Самая грубая и распространённая поправка к требуемым уровням значимости. Уровни значимости в каждом из m сравнений в m раз меньше, чем уровень значимости, требуемый при единственном сравнении. Проще говоря, делят уровень значимости ɑ на число гипотез:


```python
bonf_alpha = alhpa/16
bonf_alpha
```

Мы получили скорректированный уровень значимости на множественные сравнения (16 тестов). Повторно найдём вероятность того, что хотя бы в одном из четырёх попарных сравнений будет зафиксирован ложнопозитивный результат,применим формулу `1-(1- alpha)**n`.

```python
FWER = 1-(1-bonf_alpha)**16

if FWER < alhpa:
    print(f'Групповая вероятность ошибки первого рода {FWER}')
    print(f'Выбранный уровень стат. значимости {bonf_alpha} оптимальный')
else:
    print(f'Групповая вероятность ошибки первого рода {FWER}')
    print(f'Выбранный уровень стат. значимости {bonf_alpha} не оптимальный,есть необходимость в снижении вероятности ложнопозитивного результата')
```

Мы нашли необходимый уровень стат. значимости для выполнения множественных тестов.Перепроведём тестирование с скорректированным уровнем стат. значимости по  всем проведённым ранее тестам.

**Анализ результатов эксперимента**

Нами были проведены A/A и A/B тесты,чтобы подтвердить или опровергнуть гипотезу о существовании либо отсутствии различий конверсий по событиям в группах. А конкретно, нас интересовало, повлияет ли изменение шрифта на пользователей.Для проведения тестов мы воспользовались стат. критерием о проверке гипотезы о равенстве долей(z тест).

Мы провели 16 стат. тестов без поправки на множественные сравнения и 16 тестов с поправкой.Результат по всем тестам категоричный: `не отвергаем H0, конверсии групп равны`. Значит мы можем утверждать, что конверсии контрольных групп с тестируемой по всем событиям равны. **Изменение шрифта не повлияло на пользователей**.

Однако относительная разница между конверсиями по всем событиям всё же демонстрирует небольшой разрыв по сравниваемым группам:
- 246 и 247
- 246 и 248
- 246+247 и 248

Конверсия всех событий по тестируемой группе меньше контрольной.Скорее всего временной диапазон данных  равный почти 7-ми дням недостаточен для стабилизации значений конверсии по группам, данные не успели накопиться.
